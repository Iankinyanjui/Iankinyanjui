```python
import pandas as pd

df = pd.read_csv("Road.csv")
print("Data loaded succefully!")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Encode the 'Severity' column to numerical values before grouping
severity_mapping = {'Slight Injury': 0, 'Serious Injury': 1, 'Fatal injury': 2}
df['Severity_encoded'] = df['Severity'].map(severity_mapping)

df.groupby('Weather')['Severity_encoded'].mean().plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Weather Condition')
plt.ylabel('Average Severity (Encoded)')
plt.title('Average Accident Severity by Weather')
plt.show()
```

```python
df.rename(columns={
    'Weather_conditions': 'Weather',
    'Light_conditions': 'Light',
    'Road_surface_conditions': 'Road_surface',
    'Age_band_of_driver' : 'Age_driver',
    'Driving_experience' : 'Driving_experience',
    'Accident_severity' : 'Severity'

}, inplace=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import pandas as pd

x = df[['Weather', 'Light', 'Road_surface', 'Age_driver', 'Driving_experience']]
y = df['Severity']


x = pd.get_dummies(x, columns=['Weather', 'Light', 'Road_surface', 'Age_driver', 'Driving_experience'], drop_first=True)

severity_mapping = {'Slight Injury': 0, 'Serious Injury': 1, 'Fatal injury': 2} # Assuming 'Fatal injury' might also be present, based on common accident severity levels.
y = y.map(severity_mapping)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print ("Model trained successfully")
print ("RÂ² score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
import joblib
joblib.dump(model, 'road_accident_severity_model.pkl')
print("ðŸ’¾ Model saved as road_accident_severity_model.pkl")

new_data = pd.DataFrame([[0] * len(x.columns)], columns=x.columns)

print("Predicted Accident Severity:", model.predict(new_data)[0])
```
